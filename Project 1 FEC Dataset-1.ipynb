```python
kc_data['party_affiliation'] = kc_data['committee_name'].map(party_affiliation)

# Create a pivot table
pivot_table_data = kc_data.pivot_table(
    values='contribution_receipt_amount',
    index='contributor_occupation',
    columns='party_affiliation',
    aggfunc='sum',
    fill_value=0
)

# Filter the pivot table
filtered_pivot_table = pivot_table_data[(pivot_table_data['Democratic'] > 2000) | (pivot_table_data['Republican'] > 2000)]

# Plot the filtered pivot table
filtered_pivot_table.plot(kind='bar', figsize=(12, 6), title='Contribution Amounts by Occupation and Party Affiliation')
plt.xlabel('Occupation')
plt.ylabel('Contribution Amount')
plt.legend(title='Party Affiliation')
plt.show()
```
